The graph below is where things get interesting. On the left, we have the average player's height in the NBA from 1996-2022, and on the right is the average player's weight. The weight is measured in kilograms, and the height is measured in cm. So, in the 2021-22 season, the average NBA player's weight was roughly 214.6 pounds, and the average height was approximately 6 feet 5 inches. From this graph, we see that from 1996-2022, the average player's weight has dropped roughly 7.6 pounds which is a good amount. When we also look at the average height from 1996-2022, that has fallen by about an inch. Especially since the 2015-2016 season, there's been about a 6.8-pound drop for the average NBA player. This is where most of the weight drop occurred, which is in the last seven years. The reason I point this out is that it is not a coincidence. 

The average NBA player has gotten smaller since teams have shifted towards smaller lineups. The NBA has focused on dropping the idea of a true big man because the game is all about shooting now. It's gotten to the point where seeing Power Forward/Centers shooting 3's isn't rare when it used to be case. For those who know the NBA, this 2015-2016 season is important because it's the season the Golden State Warriors went 73-9 and had the best regular season record of all time. Much of this has to do with how the Warriors focused on having a smaller lineup and being more hyper-focused on really moving the ball, which requires lighter and sometimes smaller players to keep up with the needed stamina. Stephen Curry changed the game and led the Warriors to this historic 73-9 season and has become the greatest shooter ever in regards to not only how much he's single-handedly changed the game to a more shooting-focused playstyle, but also the fact that he has the most three's ever made while he still has a decent bit of his career ahead of him.

```python
plt.figure(figsize=(15,5))

seasons = list(data['season'].unique())
earliest_season = seasons[0]
latest_season = seasons[-1]

sns.kdeplot(data[data['season'] == earliest_season]['pts'], color='b', label='1996', linewidth=4)
sns.kdeplot(data[data['season'] == latest_season]['pts'], color='r', label='2021', linewidth= 4)

plt.title("1996 vs 2021: Points-Per-Game Distribution", fontsize=22)
    
plt.legend()
```

The graph above compares the average PPG of the 1996-97 and 2021-22 seasons. The graph for the most recent season is shifted more to the right, which means the average PPG for players has risen. It used to be a novelty to see a player average 30 PPG, but as we can see from the graph, it's a fair bit above the 1996 line once they cross the 30 points mark on the x-axis. Even when looking at where most of the density is for these two seasons, we see it's heavily around the 4 PPG mark for the 1996-97 season and around 5 PPG for the 2021-22 season. The game has become more offensive-oriented to the point where critics have questioned if NBA teams' defenses are getting worse or if their offenses are getting better. Seeing the scoring talent nowadays, which we will look at more in a bit, it's evident that when it comes to looking at stats, NBA teams' offensive games are getting better to the point where defenses seem inadequate when in reality, they aren't getting worse. Players are just getting better at shooting the ball.

```python
plyr_21_22 =  data[data["season"] == '2021-22']
plyr_21_22 = plyr_21_22.sort_values("pts",ascending = False)
plyr_21_22 = plyr_21_22[:15]
```

```python
plot1 = sns.barplot(x = "pts", y ="player_name",data = plyr_21_22, order = plyr_21_22.sort_values("pts").player_name, 
            palette = 'bright')
plot1.set_xlim(0,40)
plot1.set_xticks(range(0,41,2))
plt.ylabel('Player Name')
plt.xlabel('Points')
plt.show()
```

Above is a graph of the top 15 scorers for the 2021-22 season. I wanted to use the most recent season as this is an example of how it's commonplace to have multiple players scoring almost at or above 30 PPG in a season. Joel Embiid led the season at 30.6 PPG, with LeBron James close behind at 30.3 PPG. As I mentioned earlier, the average NBA player has been trending towards the lighter side in weight and is slightly smaller at a drop of an inch in the past roughly seven seasons. When we look at the top 15 scorers from this past season, we see a different trend at the top of this PPG list. Joel Embiid is 7'0 tall, which for the NBA is quite a bit above average, at roughly 6'5. Then we see LeBron, who is 6'9, Kevin Durant at 6'10, and Giannis Antetokounmpo at 6'11. So if the NBA is trending downward in height and weight, why do we at least have the top four PPG leaders for this most recent season at the top of the list, well above the average height? All of these players already had the height aspect, which makes scoring in the post easier since they're taller than the average NBA player. As I mentioned before, along with the height and having an easier time in the paint, the game is moving towards a more shooting-oriented playstyle. 

When you combine a tall player with the ability to shoot the ball reasonably well, this list is what you get. These four players, especially Durant, are some of the league's best shooters, but these four, which again are at the top of the list, have both height and shooting ability. The other players down the list, in terms of a pure shooting perspective, are better than some of these top four players because they need to be. Curry is a PG and needs to be able to shoot; Booker and Irving also need to be able to shoot, and so do Trae Young and Luka Doncic, who are right behind these top-four titans. The thing is, though, these players I just mentioned are either Point Guards or Shooting Guards. If they couldn't shoot, they don't have the height to offset that lack of talent in their game. Embiid, James, Durant, and Antetokounmpo have both, which is where the top NBA players are trending towards. Height and shooting ability.

One last interesting piece of data was the average age of NBA players from 1996 to now. It's trended downward quite a bit, from 28 years old in the 1996-97 season to the 2020-21 season at just over 26 years old. That's a whole two-year difference which is quite a lot. In terms of there being a concrete answer for this, I don't believe there is one. My interpretation of it and what I believe has caused the average age to steadily decline over 25 years up until this most recent 2021-22 season is that coaches are focusing more on young talent and building teams up from them. I also think teams after that 2015-16 historic season by Golden State tried to trade for smaller players who are more shooting focused, while also trying to build up younger talent for the future—so playing more for the future instead of the present. A lot of teams gave up essentially, especially the season after, from 2016-17, when Kevin Durant joined the Golden State Warriors. They just had a historic 73-9 season, and although they did lose in the playoffs, they also acquired one of the best players in the league that off-season. 

For the next couple of years, Golden State dominated to the point where they had zero competition. Teams saw this while knowing they had no hope of winning, and I believe they tried and copied the Warrior's blueprint of being a run-and-gun shooting-oriented team. This meant acquiring younger talent leading older players to be unable to keep up literally and figuratively with the NBA's new direction. That's where you see that drop off in age which is after that 2015-16 season. I guess my point in all of this, which is shown both in the data and outside of it, is that Stephen Curry and the Warriors changed the way the game is played in almost every aspect to the point where we've been able to digest that season onward with six seasons of data. The ripple effect that the Warriors team had on the league for the composition and playstyle of the average NBA player has become the byproduct of a massively changed league and what it takes to compete in it.
